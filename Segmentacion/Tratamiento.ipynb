```python
# ============================================
# Detección de parásito con SLIC + Features
# ============================================
# - Normaliza imagen con referencia
# - SLIC superpíxeles
# - Features por píxel: LAB, LCH, grad(L), banco de Gabor
# - Agrega features por superpíxel (media y std)
# - KMeans sobre superpíxeles -> máscara candidata
# - Post-proceso morfológico y filtros de forma
# - Visualización y guardado de resultados
# ============================================

from pathlib import Path
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd

from skimage import color as skcolor, img_as_ubyte
from skimage.segmentation import slic
from skimage.filters import gabor
from skimage.measure import label, regionprops, regionprops_table
from skimage.morphology import remove_small_objects, remove_small_holes, binary_opening, binary_closing, disk
from sklearn.cluster import KMeans

# ------------------------
# Parámetros principales
# ------------------------
ROOT = "./img"                   # Carpeta base
K_SUPERPIXELS = 600              # N° aprox. de superpíxeles (ajusta según tamaño imagen)
COMPACTNESS = 10.0               # Compacidad SLIC (↑ = respeta más color, ↓ = más forma)
SIGMA_SLIC = 1.0                 # Suavizado previo en SLIC

# Banco de Gabor (frecuencias y orientaciones)
GABOR_FREQUENCIES = [0.1, 0.2, 0.3]      # c/px aprox
GABOR_THETAS = [0, np.pi/6, np.pi/3, np.pi/2, 2*np.pi/3, 5*np.pi/6]

# KMeans para superpíxeles
K_CLUSTERS = 3

# Filtros de forma (post-clustering)
MIN_AREA = 80              # px, ajusta según escala
MAX_AREA = 100000          # px
MIN_CIRCULARITY = 0.10     # 4πA/P^2
MAX_CIRCULARITY = 0.95
MAX_ELONGATION = 8.0       # major_axis/minor_axis (↓ más redondo, ↑ más alargado)

# Post-proceso morfológico
OPENING_RADIUS = 2
CLOSING_RADIUS = 2
HOLE_MIN_SIZE = 64
COMP_MIN_SIZE = 64

# ------------------------
# Utilidades
# ------------------------
def read_rgb_float01(path: str) -> np.ndarray:
    """cv2.imread -> RGB float32 en [0,1]."""
    arr = cv2.imread(path, cv2.IMREAD_COLOR)
    if arr is None:
        raise FileNotFoundError(f"No se puede leer: {path}")
    arr = cv2.cvtColor(arr, cv2.COLOR_BGR2RGB)
    return (arr.astype(np.float32) / 255.0)

def normalize_img(img: np.ndarray, rimg: np.ndarray, eps: float = 1e-8) -> np.ndarray:
    """Normaliza por imagen de referencia, mantiene en [0,1]."""
    norm = img / (rimg + eps)
    norm = np.nan_to_num(norm, nan=0.0, posinf=0.0, neginf=0.0)
    return np.clip(norm, 0.0, 1.0).astype(np.float32)

def gradients_L(lab: np.ndarray) -> np.ndarray:
    """Magnitud de gradiente Sobel sobre L* (normalizada 0-1)."""
    L = lab[..., 0]
    Ln = (L - L.min()) / (L.max() - L.min() + 1e-8)
    gx = cv2.Sobel(Ln.astype(np.float32), cv2.CV_32F, 1, 0, ksize=3)
    gy = cv2.Sobel(Ln.astype(np.float32), cv2.CV_32F, 0, 1, ksize=3)
    grad = np.sqrt(gx*gx + gy*gy)
    return grad

def gabor_bank(img_gray: np.ndarray, freqs, thetas):
    """Banco de filtros de Gabor sobre imagen [0,1] (usar L*). Devuelve lista de magnitudes."""
    feats = []
    for f in freqs:
        for t in thetas:
            real, imag = gabor(img_gray, frequency=f, theta=t)
            mag = np.sqrt(real**2 + imag**2)
            feats.append(mag.astype(np.float32))
    return feats  # lista de (H,W)

def build_pixel_features(rgb01: np.ndarray):
    """Construye features por píxel: LAB, LCH (C*,h), grad(L), Gabor(L*)."""
    lab = skcolor.rgb2lab(rgb01)          # float64
    lch = skcolor.lab2lch(lab)
    L = lab[..., 0].astype(np.float32)
    a = lab[..., 1].astype(np.float32)
    b = lab[..., 2].astype(np.float32)
    C = lch[..., 1].astype(np.float32)
    h = lch[..., 2].astype(np.float32)

    # Gradiente en L*
    gradL = gradients_L(lab).astype(np.float32)

    # Banco de Gabor en L* (normalizado a [0,1])
    Ln = (L - L.min()) / (L.max() - L.min() + 1e-8)
    gabs = gabor_bank(Ln, GABOR_FREQUENCIES, GABOR_THETAS)

    # Stack de features por píxel
    # Orden: [L, a, b, C, h, gradL, gabor_...]
    feat_list = [L, a, b, C, h, gradL] + gabs
    feats = np.stack(feat_list, axis=-1)  # (H,W,F)
    return feats, lab

def superpixel_aggregate(features: np.ndarray, segments: np.ndarray):
    """
    Agrega por superpíxel: media y desviación estándar de cada feature.
    Devuelve matriz (N_spx, 2*F) y un DataFrame auxiliar con 'superpixel' y 'n_pixels'.
    """
    H, W, F = features.shape
    spx_ids = np.unique(segments)
    N = spx_ids.size

    # aplanar
    X = features.reshape(-1, F)
    s = segments.reshape(-1)

    means = np.zeros((N, F), dtype=np.float32)
    stds  = np.zeros((N, F), dtype=np.float32)
    sizes = np.zeros((N,), dtype=np.int32)

    for i, sp_id in enumerate(spx_ids):
        mask = (s == sp_id)
        Xi = X[mask]
        means[i] = Xi.mean(axis=0)
        stds[i]  = Xi.std(axis=0)
        sizes[i] = mask.sum()

    agg = np.concatenate([means, stds], axis=1)   # (N, 2F)
    info = pd.DataFrame({"superpixel": spx_ids, "n_pixels": sizes})
    return agg, info

def kmeans_superpixels(agg_feats: np.ndarray, K: int):
    """Clustering KMeans en el espacio de features agregadas de superpíxeles."""
    km = KMeans(n_clusters=K, n_init=20, random_state=42)
    labels = km.fit_predict(agg_feats.astype(np.float32))
    centers = km.cluster_centers_
    return labels, centers

def superpixels_to_mask(spx_labels: np.ndarray, spx_ids: np.ndarray, target_label: int, segments: np.ndarray):
    """Convierte etiquetas de superpíxeles a máscara binaria (target_label==1)."""
    mapping = {sid: (1 if lbl == target_label else 0) for sid, lbl in zip(spx_ids, spx_labels)}
    map_arr = np.vectorize(mapping.get)(segments)
    return map_arr.astype(bool)

def postprocess_mask(mask: np.ndarray) -> np.ndarray:
    """Morfología: limpieza de objetos pequeños, apertura/cierre y relleno de huecos."""
    mask = remove_small_objects(mask, min_size=COMP_MIN_SIZE)
    mask = binary_opening(mask, footprint=disk(OPENING_RADIUS))
    mask = binary_closing(mask, footprint=disk(CLOSING_RADIUS))
    mask = remove_small_holes(mask, area_threshold=HOLE_MIN_SIZE)
    return mask

def shape_filter(mask: np.ndarray,
                 min_area=MIN_AREA, max_area=MAX_AREA,
                 min_circ=MIN_CIRCULARITY, max_circ=MAX_CIRCULARITY,
                 max_elong=MAX_ELONGATION) -> np.ndarray:
    """
    Filtros de forma por componente:
    - área en [min_area, max_area]
    - circularidad = 4πA/P^2 en [min_circ, max_circ]
    - elongación = major/minor <= max_elong
    Devuelve una máscara con sólo las componentes válidas.
    """
    lab_img = label(mask)
    out = np.zeros_like(mask, dtype=bool)

    for r in regionprops(lab_img):
        A = r.area
        if r.perimeter <= 0:
            circ = 0.0
        else:
            circ = 4.0 * np.pi * A / (r.perimeter ** 2)

        # Ejes principales (si no hay minor_axis_length suficientemente grande, evitar división)
        maj = getattr(r, "major_axis_length", 0.0) or 0.0
        minax = getattr(r, "minor_axis_length", 0.0) or 1.0
        elong = (maj / minax) if minax > 0 else np.inf

        ok = (min_area <= A <= max_area) and (min_circ <= circ <= max_circ) and (elong <= max_elong)
        if ok:
            out[lab_img == r.label] = True

    return out

def overlay_mask_on_rgb(rgb01: np.ndarray, mask: np.ndarray, alpha=0.45):
    """Pinta la máscara en RGB (tono magenta)."""
    overlay = rgb01.copy()
    color = np.array([1.0, 0.0, 1.0], dtype=np.float32)  # magenta
    overlay[mask] = (1 - alpha) * overlay[mask] + alpha * color
    return overlay

# ------------------------
# Flujo principal por par imagen/ref
# ------------------------
def process_pair(img_path: Path, ref_path: Path, out_dir: Path):
    out_dir.mkdir(parents=True, exist_ok=True)

    # 1) Lectura y normalización
    rgb = read_rgb_float01(str(img_path))
    rimg = read_rgb_float01(str(ref_path))
    norm = normalize_img(rgb, rimg)

    # 2) Features por píxel
    feats_px, lab = build_pixel_features(norm)
    H, W, F = feats_px.shape

    # 3) SLIC superpíxeles
    segments = slic(
        norm,
        n_segments=K_SUPERPIXELS,
        compactness=COMPACTNESS,
        sigma=SIGMA_SLIC,
        start_label=0,
        channel_axis=-1
    )

    # 4) Agregación por superpíxel
    agg_feats, spx_info = superpixel_aggregate(feats_px, segments)  # (N_spx, 2F), DataFrame
    spx_ids = spx_info["superpixel"].to_numpy()

    # 5) KMeans en superpíxeles
    spx_labels, spx_centers = kmeans_superpixels(agg_feats, K=K_CLUSTERS)

    # 6) Heurística: escoger cluster candidato al parásito
    #    Idea simple: el parásito (morado) tiende a tener C* alto (más cromático)
    #    Calculamos media de C* (feature índice 3 en las medias) por cluster y cogemos el mayor.
    Fhalf = feats_px.shape[-1]             # número de features por píxel
    # agg_feats = [means | stds] -> índice 3 corresponde a C* en medios
    Cstar_means = []
    for k in range(K_CLUSTERS):
        Ck = agg_feats[spx_labels == k, 3]  # columna 3 = mean(C*)
        Cstar_means.append(Ck.mean() if Ck.size > 0 else -np.inf)
    target_label = int(np.argmax(Cstar_means))

    # 7) Máscara desde superpíxeles y post-proceso
    mask0 = superpixels_to_mask(spx_labels, spx_ids, target_label, segments)
    mask1 = postprocess_mask(mask0)
    mask2 = shape_filter(mask1)  # filtra por forma

    # 8) Visualización rápida
    seg_overlay0 = overlay_mask_on_rgb(norm, mask0)
    seg_overlay2 = overlay_mask_on_rgb(norm, mask2)

    plt.figure(figsize=(12, 6))
    plt.subplot(1,3,1); plt.imshow(norm); plt.title("Normalizada"); plt.axis("off")
    plt.subplot(1,3,2); plt.imshow(seg_overlay0); plt.title("Cluster candidato (C* alto)"); plt.axis("off")
    plt.subplot(1,3,3); plt.imshow(seg_overlay2); plt.title("Tras filtros de forma"); plt.axis("off")
    plt.tight_layout(); plt.show()

    # 9) Guardados
    base = img_path.stem
    Path(out_dir / f"{base}_overlay_candidate.png").write_bytes(
        cv2.imencode(".png", cv2.cvtColor(img_as_ubyte(seg_overlay0), cv2.COLOR_RGB2BGR))[1].tobytes()
    )
    Path(out_dir / f"{base}_overlay_final.png").write_bytes(
        cv2.imencode(".png", cv2.cvtColor(img_as_ubyte(seg_overlay2), cv2.COLOR_RGB2BGR))[1].tobytes()
    )

    # CSV de features por superpíxel + etiqueta de cluster
    df_feats = pd.DataFrame(agg_feats, columns=[*(f"mean_f{i}" for i in range(F)), *(f"std_f{i}" for i in range(F))])
    # Nombres legibles para los primeros 6 features (L,a,b,C,h,gradL):
    nice = ["L","a","b","C","h","gradL"]
    for i, name in enumerate(nice):
        df_feats = df_feats.rename(columns={f"mean_f{i}": f"mean_{name}", f"std_f{i}": f"std_{name}"})
    # Añade gabor labels (opcional)
    # df_feats.rename(...) si quieres mapear mean_f6.. a mean_gabor_f{..}_theta{..}
    df_feats["superpixel"] = spx_ids
    df_feats["cluster"] = spx_labels
    df_feats["n_pixels"] = spx_info["n_pixels"].values

    # Guarda CSV
    df_feats.to_csv(out_dir / f"{base}_superpixel_features.csv", index=False)

    print(f"[OK] Guardado: {out_dir / (base + '_overlay_final.png')}")
    print(f"[OK] CSV     : {out_dir / (base + '_superpixel_features.csv')}")

# ------------------------
# Descubrimiento de pares y ejecución
# ------------------------
def find_pairs(root: str):
    pairs = []
    for ref_path in Path(root).rglob("*_ref.png"):
        img_path = Path(str(ref_path).replace("_ref.png", ".png"))
        if img_path.exists():
            pairs.append((img_path, ref_path))
    return pairs

if __name__ == "__main__":
    out_dir = Path(ROOT) / "out_parasite"
    pairs = find_pairs(ROOT)
    print(f"Pares encontrados: {len(pairs)}")
    if not pairs:
        print("No se encontraron pares *_ref.png. Revisa rutas.")
    for img_path, ref_path in pairs:
        print(f"Procesando: {img_path.name}  (ref: {ref_path.name})")
        process_pair(img_path, ref_path, out_dir)
```
