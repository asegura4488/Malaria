```python
# ============================================================
# ENTRENAMIENTO (./img) + TESTING (./img_testing) NO SUPERVISADO
# SLIC + features (LAB/LCH/grad/Gabor) + KMeans
# CON: tagging de "parásito" (cluster con mayor C*) y rótulo en imagen,
#      resúmenes por imagen y gráficos comparativos.
# ============================================================

from pathlib import Path
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import joblib

from skimage import color as skcolor, img_as_ubyte
from skimage.segmentation import slic, mark_boundaries
from skimage.filters import gabor
from skimage.morphology import remove_small_objects, remove_small_holes, binary_opening, binary_closing, disk
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# ------------------ Configuración ------------------
ROOT_TRAIN = "./img"
ROOT_TEST  = "./img_testing"

OUT_TRAIN = Path(ROOT_TRAIN) / "out_unsup_train"
OUT_TEST  = Path(ROOT_TEST)  / "out_unsup_test"
OUT_TRAIN.mkdir(parents=True, exist_ok=True)
OUT_TEST.mkdir(parents=True, exist_ok=True)

MODEL_PATH = "model_unsup.joblib"

# SLIC
K_SUPERPIXELS = 600
COMPACTNESS = 10.0
SIGMA_SLIC = 1.0

# Gabor (textura)
GABOR_FREQUENCIES = [0.1, 0.2, 0.3]
GABOR_THETAS = [0, np.pi/6, np.pi/3, np.pi/2, 2*np.pi/3, 5*np.pi/6]

# Clustering
K_CLUSTERS = 3       # Nº de clusters global

# Morfología (para heurística “parasite_by_Cstar”)
OPENING_RADIUS = 2
CLOSING_RADIUS = 2
HOLE_MIN_SIZE = 64
COMP_MIN_SIZE = 64

PARASITE_HEURISTIC = True  # Taggear clúster con mayor C* como "parásito"

# ------------------ Utilidades base ------------------
def read_rgb_float01(path: str) -> np.ndarray:
    arr = cv2.imread(path, cv2.IMREAD_COLOR)
    if arr is None:
        raise FileNotFoundError(f"No se puede leer: {path}")
    arr = cv2.cvtColor(arr, cv2.COLOR_BGR2RGB)
    return (arr.astype(np.float32) / 255.0)

def normalize_by_ref(rgb: np.ndarray, ref: np.ndarray, eps: float = 1e-8) -> np.ndarray:
    norm = rgb / (ref + eps)
    norm = np.nan_to_num(norm, nan=0.0, posinf=0.0, neginf=0.0)
    return np.clip(norm, 0.0, 1.0).astype(np.float32)

def gradients_L_from_lab(lab: np.ndarray) -> np.ndarray:
    L = lab[..., 0]
    Ln = (L - L.min()) / (L.max() - L.min() + 1e-8)
    gx = cv2.Sobel(Ln.astype(np.float32), cv2.CV_32F, 1, 0, ksize=3)
    gy = cv2.Sobel(Ln.astype(np.float32), cv2.CV_32F, 0, 1, ksize=3)
    return np.sqrt(gx*gx + gy*gy)

def gabor_bank(gray01: np.ndarray, freqs, thetas):
    feats = []
    for f in freqs:
        for t in thetas:
            real, imag = gabor(gray01, frequency=f, theta=t)
            mag = np.sqrt(real**2 + imag**2).astype(np.float32)
            feats.append(mag)
    return feats  # lista (H,W)

def pixel_features(rgb01: np.ndarray):
    lab = skcolor.rgb2lab(rgb01)
    lch = skcolor.lab2lch(lab)
    L = lab[...,0].astype(np.float32)
    a = lab[...,1].astype(np.float32)
    b = lab[...,2].astype(np.float32)
    C = lch[...,1].astype(np.float32)
    h = lch[...,2].astype(np.float32)
    gradL = gradients_L_from_lab(lab).astype(np.float32)
    Ln = (L - L.min()) / (L.max() - L.min() + 1e-8)
    gabs = gabor_bank(Ln, GABOR_FREQUENCIES, GABOR_THETAS)
    feat_list = [L, a, b, C, h, gradL] + gabs
    feats = np.stack(feat_list, axis=-1)  # (H,W,F)
    return feats, lab

def superpixel_aggregate(features_hwf: np.ndarray, segments_hw: np.ndarray):
    H, W, F = features_hwf.shape
    spx_ids = np.unique(segments_hw)
    X = features_hwf.reshape(-1, F)
    s = segments_hw.reshape(-1)
    means = np.zeros((spx_ids.size, F), dtype=np.float32)
    stds  = np.zeros_like(means)
    sizes = np.zeros((spx_ids.size,), dtype=np.int32)
    for i, sid in enumerate(spx_ids):
        mask = (s == sid)
        Xi = X[mask]
        means[i] = Xi.mean(axis=0)
        stds[i]  = Xi.std(axis=0)
        sizes[i] = mask.sum()
    agg = np.concatenate([means, stds], axis=1)  # (N_spx, 2F)
    return agg, spx_ids, sizes

def colorize_clusters(segments: np.ndarray, labels: np.ndarray, K: int) -> np.ndarray:
    H, W = segments.shape
    cmap = plt.get_cmap("tab10")
    out = np.zeros((H, W, 3), dtype=np.float32)
    for sid, lbl in zip(np.unique(segments), labels):
        out[segments == sid] = cmap(lbl % 10)[:3]
    return out

def postprocess_mask(mask: np.ndarray) -> np.ndarray:
    mask = remove_small_objects(mask, min_size=COMP_MIN_SIZE)
    mask = binary_opening(mask, footprint=disk(OPENING_RADIUS))
    mask = binary_closing(mask, footprint=disk(CLOSING_RADIUS))
    mask = remove_small_holes(mask, area_threshold=HOLE_MIN_SIZE)
    return mask

def overlay_mask(rgb01: np.ndarray, mask: np.ndarray, alpha=0.45) -> np.ndarray:
    out = rgb01.copy()
    color = np.array([1.0, 0.0, 1.0], dtype=np.float32)
    out[mask] = (1 - alpha) * out[mask] + alpha * color
    return out

def put_tag(img_rgb01: np.ndarray, text: str, pos=(10, 30)):
    """Escribe texto legible en RGB[0,1]."""
    img8 = img_as_ubyte(img_rgb01.copy())
    cv2.putText(img8, text, pos, cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,0), 3, cv2.LINE_AA)
    cv2.putText(img8, text, pos, cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,255,255), 1, cv2.LINE_AA)
    return img8  # BGR conversion se hace al guardar

def collect_images(root: str):
    exts = ["*.png", "*.jpg", "*.jpeg", "*.tif", "*.tiff", "*.bmp"]
    files = []
    for ext in exts:
        files += list(Path(root).rglob(ext))
    files = [p for p in files if not p.name.endswith("_ref.png")]
    return sorted(files)

# ------------------ Empaquetado por imagen ------------------
def load_and_prepare_image(img_path: Path):
    ref_path = Path(str(img_path).rsplit(".", 1)[0] + "_ref.png")
    rgb = read_rgb_float01(str(img_path))
    if ref_path.exists():
        ref = read_rgb_float01(str(ref_path))
        rgbn = normalize_by_ref(rgb, ref)
    else:
        rgbn = rgb
    feats_px, lab = pixel_features(rgbn)
    H, W, F = feats_px.shape
    segments = slic(
        rgbn, n_segments=K_SUPERPIXELS, compactness=COMPACTNESS,
        sigma=SIGMA_SLIC, start_label=0, channel_axis=-1
    )
    agg, spx_ids, sizes = superpixel_aggregate(feats_px, segments)
    return {"rgb": rgbn, "lab": lab, "features_px": feats_px, "segments": segments,
            "agg": agg, "spx_ids": spx_ids, "sizes": sizes, "F_each": F}

# ------------------ Métricas internas por imagen ------------------
def image_internal_metrics(X_img: np.ndarray, labels_img: np.ndarray):
    metrics = {"silhouette": np.nan, "davies_bouldin": np.nan, "calinski_harabasz": np.nan}
    if len(np.unique(labels_img)) < 2 or X_img.shape[0] < 3:
        return metrics
    try: metrics["silhouette"] = float(silhouette_score(X_img, labels_img))
    except Exception: pass
    try: metrics["davies_bouldin"] = float(davies_bouldin_score(X_img, labels_img))
    except Exception: pass
    try: metrics["calinski_harabasz"] = float(calinski_harabasz_score(X_img, labels_img))
    except Exception: pass
    return metrics

# ------------------ Pipeline TRAIN/TEST con Tagueo ------------------
def train_unsup(root_train: str):
    imgs = collect_images(root_train)
    if not imgs:
        raise RuntimeError("No se encontraron imágenes de entrenamiento en ./img")

    per_image, all_agg = [], []
    for p in imgs:
        data = load_and_prepare_image(p)
        per_image.append((p, data))
        all_agg.append(data["agg"])
        print(f"[TRAIN] {p.name}: superpix={data['agg'].shape[0]}")
    X_train_all = np.vstack(all_agg)

    scaler = StandardScaler()
    Xs = scaler.fit_transform(X_train_all.astype(np.float32))
    kmeans = KMeans(n_clusters=K_CLUSTERS, n_init=50, random_state=42)
    kmeans.fit(Xs)

    joblib.dump({"scaler": scaler, "kmeans": kmeans}, MODEL_PATH)
    print(f"[OK] Modelo guardado en {MODEL_PATH}")

    # métricas + tag por imagen
    metric_rows = []
    tag_rows = []

    for p, data in per_image:
        agg = data["agg"]; seg = data["segments"]; spx_ids = data["spx_ids"]
        sizes = data["sizes"]; rgb = data["rgb"]; F_each = data["F_each"]

        X_img = scaler.transform(agg.astype(np.float32))
        lbls = kmeans.predict(X_img)

        # Overlay por clúster
        cluster_rgb = colorize_clusters(seg, lbls, K_CLUSTERS)
        overlay_clusters = img_as_ubyte(mark_boundaries(cluster_rgb, seg))

        cv2.imwrite(str(OUT_TRAIN / f"{p.stem}_clusters_overlay.png"),
                    cv2.cvtColor(overlay_clusters, cv2.COLOR_RGB2BGR))

        # Heurística: clúster parásito por C* medio (columna 3 de los medios)
        mean_C_col = 3
        C_means = []
        for k in range(K_CLUSTERS):
            vals = agg[lbls == k, mean_C_col]
            C_means.append(vals.mean() if vals.size else -np.inf)
        parasite_label = int(np.argmax(C_means))
        parasite_meanC = C_means[parasite_label] if np.isfinite(C_means[parasite_label]) else float("nan")

        # Máscara y postproceso
        mask = np.isin(seg, spx_ids[lbls == parasite_label])
        mask_pp = postprocess_mask(mask)

        # Overlay + TAG de texto en la imagen
        overlay = overlay_mask(rgb, mask_pp, alpha=0.45)
        text = f"Parasite (heurística por C*): cluster={parasite_label}, mean C*={parasite_meanC:.2f}"
        overlay_tag = put_tag(overlay, text, pos=(10, 30))
        cv2.imwrite(str(OUT_TRAIN / f"{p.stem}_parasite_overlay.png"),
                    cv2.cvtColor(overlay_tag, cv2.COLOR_RGB2BGR))

        # CSV por imagen con features + cluster
        nice = ["L","a","b","C","h","gradL"]
        n_gab = F_each - len(nice)
        col_means = [f"mean_{n}" for n in nice] + [f"mean_gabor_{i}" for i in range(n_gab)]
        col_stds  = [f"std_{n}"  for n in nice] + [f"std_gabor_{i}"  for i in range(n_gab)]
        cols = col_means + col_stds
        df = pd.DataFrame(agg, columns=cols)
        df["superpixel"] = spx_ids
        df["cluster"] = lbls
        df.to_csv(OUT_TRAIN / f"{p.stem}_superpixel_features_kmeans.csv", index=False)

        # Métricas internas (en espacio escalado)
        m = image_internal_metrics(X_img, lbls)
        m["image"] = p.name
        metric_rows.append(m)

        # Resumen de tag por imagen
        spx_par = int(np.sum(lbls == parasite_label))
        pix_par = int(np.sum(sizes[lbls == parasite_label]))
        pix_tot = int(np.sum(sizes))
        frac_par = pix_par / max(1, pix_tot)
        tag_rows.append({
            "image": p.name,
            "parasite_cluster": parasite_label,
            "parasite_mean_C": parasite_meanC,
            "parasite_superpixels": spx_par,
            "total_superpixels": int(len(spx_ids)),
            "parasite_pixels": pix_par,
            "total_pixels": pix_tot,
            "parasite_area_fraction": frac_par
        })

    pd.DataFrame(metric_rows)[["image","silhouette","davies_bouldin","calinski_harabasz"]] \
      .to_csv(OUT_TRAIN / "metrics_train.csv", index=False)
    pd.DataFrame(tag_rows).to_csv(OUT_TRAIN / "tag_summary_train.csv", index=False)
    print(f"[OK] Métricas y tag summary (train) en {OUT_TRAIN}")

def test_unsup(root_test: str):
    imgs = collect_images(root_test)
    if not imgs:
        raise RuntimeError("No se encontraron imágenes de testing en ./img_testing")

    bundle = joblib.load(MODEL_PATH)
    scaler = bundle["scaler"]; kmeans = bundle["kmeans"]

    metric_rows = []
    tag_rows = []

    for p in imgs:
        data = load_and_prepare_image(p)
        agg = data["agg"]; seg = data["segments"]; spx_ids = data["spx_ids"]
        sizes = data["sizes"]; rgb = data["rgb"]; F_each = data["F_each"]

        X_img = scaler.transform(agg.astype(np.float32))
        lbls = kmeans.predict(X_img)

        # Overlays
        cluster_rgb = colorize_clusters(seg, lbls, K_CLUSTERS)
        overlay_clusters = img_as_ubyte(mark_boundaries(cluster_rgb, seg))
        cv2.imwrite(str(OUT_TEST / f"{p.stem}_clusters_overlay.png"),
                    cv2.cvtColor(overlay_clusters, cv2.COLOR_RGB2BGR))

        mean_C_col = 3
        C_means = []
        for k in range(K_CLUSTERS):
            vals = agg[lbls == k, mean_C_col]
            C_means.append(vals.mean() if vals.size else -np.inf)
        parasite_label = int(np.argmax(C_means))
        parasite_meanC = C_means[parasite_label] if np.isfinite(C_means[parasite_label]) else float("nan")

        mask = np.isin(seg, spx_ids[lbls == parasite_label])
        mask_pp = postprocess_mask(mask)
        overlay = overlay_mask(rgb, mask_pp, alpha=0.45)
        text = f"Parasite (heurística por C*): cluster={parasite_label}, mean C*={parasite_meanC:.2f}"
        overlay_tag = put_tag(overlay, text, pos=(10, 30))
        cv2.imwrite(str(OUT_TEST / f"{p.stem}_parasite_overlay.png"),
                    cv2.cvtColor(overlay_tag, cv2.COLOR_RGB2BGR))

        # CSV por imagen
        nice = ["L","a","b","C","h","gradL"]
        n_gab = F_each - len(nice)
        col_means = [f"mean_{n}" for n in nice] + [f"mean_gabor_{i}" for i in range(n_gab)]
        col_stds  = [f"std_{n}"  for n in nice] + [f"std_gabor_{i}"  for i in range(n_gab)]
        cols = col_means + col_stds
        df = pd.DataFrame(agg, columns=cols)
        df["superpixel"] = spx_ids
        df["cluster"] = lbls
        df.to_csv(OUT_TEST / f"{p.stem}_superpixel_features_kmeans.csv", index=False)

        # Métricas internas (en espacio escalado)
        m = image_internal_metrics(X_img, lbls)
        m["image"] = p.name
        metric_rows.append(m)

        # Resumen de tag por imagen
        spx_par = int(np.sum(lbls == parasite_label))
        pix_par = int(np.sum(sizes[lbls == parasite_label]))
        pix_tot = int(np.sum(sizes))
        frac_par = pix_par / max(1, pix_tot)
        tag_rows.append({
            "image": p.name,
            "parasite_cluster": parasite_label,
            "parasite_mean_C": parasite_meanC,
            "parasite_superpixels": spx_par,
            "total_superpixels": int(len(spx_ids)),
            "parasite_pixels": pix_par,
            "total_pixels": pix_tot,
            "parasite_area_fraction": frac_par
        })

    pd.DataFrame(metric_rows)[["image","silhouette","davies_bouldin","calinski_harabasz"]] \
      .to_csv(OUT_TEST / "metrics_test.csv", index=False)
    pd.DataFrame(tag_rows).to_csv(OUT_TEST / "tag_summary_test.csv", index=False)
    print(f"[OK] Métricas y tag summary (test) en {OUT_TEST}")

# ------------------ Gráficas comparativas del tag ------------------
def plot_tag_comparison():
    train_csv = OUT_TRAIN / "tag_summary_train.csv"
    test_csv  = OUT_TEST / "tag_summary_test.csv"
    if not train_csv.exists() or not test_csv.exists():
        print("No se encontraron tag_summary_* para graficar.")
        return
    df_tr = pd.read_csv(train_csv)
    df_te = pd.read_csv(test_csv)
    df_tr["split"] = "train"; df_te["split"] = "test"
    df = pd.concat([df_tr, df_te], ignore_index=True)

    # Barras: porcentaje de área etiquetada como "parásito"
    plt.figure(figsize=(7,5))
    means = df.groupby("split")["parasite_area_fraction"].mean()
    means.plot(kind="bar")
    plt.title("Fracción de área (px) taggeada como 'parásito' por split")
    plt.ylabel("Fracción de área")
    plt.ylim(0, 1)
    plt.tight_layout()
    plt.savefig("tag_area_fraction_train_vs_test.png", dpi=160)
    plt.show()

    # Dispersión por imagen
    for split in ["train", "test"]:
        d = df[df["split"] == split].sort_values("parasite_area_fraction")
        plt.figure(figsize=(9,4))
        x = np.arange(len(d))
        plt.scatter(x, d["parasite_area_fraction"], s=20)
        plt.title(f"Fracción de área 'parásito' por imagen ({split})")
        plt.xlabel("Imágenes (ordenadas)")
        plt.ylabel("Fracción de área")
        plt.ylim(0, 1)
        plt.tight_layout()
        plt.savefig(f"tag_area_fraction_{split}.png", dpi=160)
        plt.show()

# ------------------ Main ------------------
if __name__ == "__main__":
    print("=== Entrenando con ./img ===")
    train_unsup(ROOT_TRAIN)
    print("=== Probando con ./img_testing ===")
    test_unsup(ROOT_TEST)
    print("=== Graficando comparación de tags (área) ===")
    plot_tag_comparison()
    print("Listo ✅")
```

```python
# ============================================
# Etiquetado a nivel de imagen (sí/no parásito)
# a partir de parasite_area_fraction
# ============================================
import os
import pandas as pd

TRAIN_SUM = "./img/out_unsup_train/tag_summary_train.csv"
TEST_SUM  = "./img_testing/out_unsup_test/tag_summary_test.csv"

# umbral configurable (ej. 0.05 = 5% del área)
THRESHOLD = 0.1

def tag_images(csv_path, threshold=THRESHOLD):
    if not os.path.exists(csv_path):
        print(f"⚠️ No encontrado: {csv_path}")
        return None
    df = pd.read_csv(csv_path)
    if "parasite_area_fraction" not in df.columns:
        raise ValueError("Falta columna 'parasite_area_fraction'")

    # 1) etiqueta por “cualquier área > 0”
    df["image_tag_any"] = (df["parasite_area_fraction"] > 0).map({True:"parasite", False:"no_parasite"})

    # 2) etiqueta por umbral
    df["image_tag_thresh"] = (df["parasite_area_fraction"] >= threshold).map({True:"parasite", False:"no_parasite"})

    # guardar
    out_csv = csv_path.replace(".csv", f"_with_image_tags_thr{threshold:.3f}.csv")
    df.to_csv(out_csv, index=False)

    # resumen
    print(f"\n{os.path.basename(out_csv)}")
    print("— Regla ANY      :", df["image_tag_any"].value_counts().to_dict())
    print(f"— Regla THRESH={threshold:.3f} :", df["image_tag_thresh"].value_counts().to_dict())
    if (df["image_tag_any"] == "parasite").all():
        print("✔ Todas las imágenes fueron marcadas como 'parasite' por la regla ANY (coincide con tu observación).")
    return df

if __name__ == "__main__":
    print("Etiquetando TRAIN:")
    df_tr = tag_images(TRAIN_SUM, THRESHOLD)
    print("\nEtiquetando TEST:")
    df_te = tag_images(TEST_SUM, THRESHOLD)
```
