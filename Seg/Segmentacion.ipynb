```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def apply_kmeans_segmentation(image, n_clusters=3):
    """
    Aplica K-Means al espacio HSV de una imagen para segmentarla en n_clusters.
    """
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    pixels = hsv_image.reshape(-1, 3)
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(pixels)
    clustered_image = kmeans.labels_.reshape(hsv_image.shape[:2])
    return clustered_image, kmeans.cluster_centers_

# Rutas de las imgenes
image_paths = [
    "000_seg_000.png",
    "000_seg_001.png",
    "000_seg_009.png",
    "000_seg_010.png",
    "000_seg_012.png",
    "000_seg_018.png",
]

# Leer imgenes
images = [cv2.imread(image_path) for image_path in image_paths]

# Aplicar K-Means a las imgenes
segmented_image_paths = []
mask_image_paths = []

for i, img in enumerate(images):
    clustered_img, _ = apply_kmeans_segmentation(img, n_clusters=3)

    # Guardar la imagen segmentada
    segmented_path = f"segmentation_result_{i+1}.png"
    plt.imsave(segmented_path, clustered_img, cmap='viridis')
    segmented_image_paths.append(segmented_path)

    # Guardar mscaras binarias
    cluster_paths = []
    for cluster_id in range(3):
        mask = (clustered_img == cluster_id).astype(np.uint8) * 255
        mask_path = f"mask_cluster_{cluster_id+1}_image_{i+1}.png"
        cv2.imwrite(mask_path, mask)
        cluster_paths.append(mask_path)
    mask_image_paths.append(cluster_paths)

# Imprimir las rutas de las imgenes y mscaras generadas
print("Imgenes segmentadas:")
for path in segmented_image_paths:
    print(path)

print("Mscaras generadas:")
for paths in mask_image_paths:
    for path in paths:
        print(path)
```
